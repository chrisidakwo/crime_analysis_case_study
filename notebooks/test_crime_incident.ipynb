#### Use a list comprehension to get all violent incidents

#### Use a list comprehension to get all incidents that resulted in an arrest

```python
resolved_incidents = [i for i in incidents if i.was_resolved()]

print(f"\n INCIDENTS WITH ARRESTS: {len(resolved_incidents)} found")
for incident in resolved_incidents:
    print(f"    {incident}")
```

```python
total_incidents = len(incidents)
```

```python
# 1. What percentage of incidents are violent?
violent_count = len(violent_incidents)
violent_pct = (violent_count / total_incidents) * 100
print(f"\n 1. Violent crime percentage: {violent_pct:.1f}% ({violent_count} out of {total_incidents})")

# 2. What is the arrest rate across incidents?
arrested_count = len(resolved_incidents)
arrest_rate = (arrested_count / total_incidents) * 100
print(f"\n 2. Arrest rate: {arrest_rate:.1f}% ({arrested_count} out of {total_incidents})")

# 3. How many unique districts appear across incidents?
unique_districts = set(i.district for i in incidents)
print(f"\n 3. Unique districts: {len(unique_districts)} -> {sorted(unique_districts)}")

# 4. Which crime category has the highest arrest rate among incidents?
categories_count = {}

for incident in incidents:
    _i_crime_category = incident.crime_category()

    _category_value = categories_count.get(_i_crime_category)

    if _category_value is not None:
        categories_count[_i_crime_category] = _category_value + incident.arrest
    else:
        categories_count[_i_crime_category] = incident.arrest

print(f"\n 4. Crime category with the highest arrest rate: {max(categories_count, key=categories_count.get)} -> {categories_count}")
```
