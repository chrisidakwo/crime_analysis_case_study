```python
incident_data = [
    (100000001, "THEFT", "2022-01-05 14:00:00", "008", 0, 0),
    (100000002, "BATTERY", "2022-01-06 22:15:00", "011", 1, 1),
    (100000003, "HOMICIDE", "2022-01-07 03:30:00", "007", 0, 0),
    (100000004, "ROBBERY", "2022-01-08 19:45:00", "011", 0, 0),
    (100000005, "THEFT", "2022-01-09 11:00:00", "014", 0, 0),
]

# Create a list of CrimeIncident objects
incidents: list[CrimeIncident] = []

for incident in incident_data:
    incidents.append(
        CrimeIncident(
            incident_id=incident[0],
            primary_type=incident[1],
            incident_date=incident[2],
            district=incident[3],
            arrest=incident[4],
            domestic=incident[5],
        )
    )

# Alternatively, you can build the content of the incidents list using list comprehension
# incidents = [
#     CrimeIncident(id_, type_, date, district, arrest, domestic)
#     for (id_, type_, date, district, arrest, domestic) in incident_data
# ]

violent_incidents = [i for i in incidents if i.is_violent()]
resolved_incidents = [i for i in incidents if i.was_resolved()]
domestic_incidents = [i for i in incidents if i.is_domestic_violence()]


# for i in incidents:
#     if i.is_violent():
#         violent_incidents.append(i)

print(f"Total incidents: {len(incidents)}")
print(f"Violent incidents: {len(violent_incidents)}")
print(f"Arrests made: {len(resolved_incidents)}")
print(f"Domestic violence: {len(domestic_incidents)}")

print("\nIncident Summaries")
for incident in incidents:
    print(f"    {incident.summary()}")

```
